# Load Data

```python
df = pd.read_pickle("/content/drive/MyDrive/ColabNotebooks/Fourth_Pass: Ale_DataBase/no_f_compounds.pkl")
df.head(5)
```

```python
print(len(df))
```

```python
df = df[df['crossings_+-1eV']<=40]
```

```python
element_list0 = list()
for i in range(1,119):
  element_list0.append(str(element(i).symbol))
```

```python
element_dict = {key: 0 for key in element_list0}
```

```python
#element_dict
```

```python
element_list1 = list()

for item_i in df.index:
  compo0 = list(df.loc[item_i, 'composition'])

  for item_j in range(len(compo0)):
    elem = element(str(compo0[item_j]))
    #print(elem.symbol)
    element_list1.append(str(elem.symbol))
```

```python
len(element_list1)
```

```python
element, counts   = np.unique(element_list1, return_counts=True)
element_list      = dict(zip(element, counts))
element_dict.update(element_list)
```

```python
element_dict
```

```python
Element_Counting = pd.DataFrame([element_dict], 
                                #columns=['count']
                                ).T
```
