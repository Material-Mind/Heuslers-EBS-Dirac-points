```python
data = df.reset_index(drop=True)
data.head(5)
```

```python
print(list(data.columns))
```

```python
f_elements_list = ['La', 'Ce', 'Pr', 'Nd', 'Pm', 'Sm', 'Eu', 'Gd', 'Tb', 'Dy', 'Ho', 'Er', 'Tm', 'Yb', 'Lu', \
                   'Ac', 'Th', 'Pa', 'U', 'Np', 'Pu', 'Am', 'Cm', 'Bk', 'Cf', 'Es', 'Fm', 'Md', 'No', 'Lr', \
                   'Ubu', 'Ubb', 'Ubt', 'Ubq', 'Ubp', 'Ubh']
```

```python
data_f_compounds    = []
data_no_f_compounds = []

for item_i in range(data.shape[0]):
  composition = list(data.loc[item_i, 'composition'])
  compo_list = []
  
  for item_j in range(len(composition)):
    compo_list.append(str(composition[item_j]))

  checking = any(elem in compo_list for elem in f_elements_list)
  if checking==True:
    data_f_compounds.append(data.iloc[item_i].values)
  else:
    data_no_f_compounds.append(data.iloc[item_i].values)
```

```python
len(data_f_compounds)
```

```python
df_f_compounds = pd.DataFrame(data=data_f_compounds, columns = list(data.columns))
df_f_compounds.head(5)
```

```python
len(data_no_f_compounds)
```

```python
df_no_f_compounds = pd.DataFrame(data = data_no_f_compounds, columns = list(data.columns))
df_no_f_compounds.head(5)
```

```python
fig = plt.figure(figsize=(8, 6))
sns.distplot(data['crossings_+-1eV'], 
             kde=False, 
             color="k", 
             label="all data", 
             hist_kws={"histtype": "stepfilled", "color":"k", "linewidth": 2, "alpha": 0.4})

sns.distplot(df_no_f_compounds['crossings_+-1eV'], 
             kde=False, 
             label="no f compounds", 
             hist_kws={"histtype": "step", "color":"b", "linewidth": 2, "alpha": 1.0})

sns.distplot(df_f_compounds['crossings_+-1eV'], 
             kde=False, 
             label="f compounds", 
             hist_kws={"histtype": "step", "color":"r", "linewidth": 2, "alpha": 1.0})


plt.legend(loc='upper right', fontsize = 16, ncol = 1).get_frame().set_edgecolor('w')
plt.tick_params(axis='x', labelsize = 15)
plt.tick_params(axis='y', labelsize = 15)
plt.xlabel('number of crossings within $\pm$ 1eV', fontsize = 20)
#plt.xscale('symlog')
plt.ylabel('Counting', fontsize = 20)
plt.yscale('linear');
```
