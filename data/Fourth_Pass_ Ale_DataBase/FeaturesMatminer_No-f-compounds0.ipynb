# **Stuctural Featurizer**

```python
structure_featurizer = MultipleFeaturizer([DensityFeatures(), 
                                           MaximumPackingEfficiency(), 
                                           EwaldEnergy(accuracy=4, per_atom=True), 
                                           #OrbitalFieldMatrix(), 
                                           ElectronicRadialDistributionFunction(), 
                                           StructuralHeterogeneity(), 
                                           StructuralComplexity(), 
                                           StructureComposition(ValenceOrbital(props=['frac'])), 
                                           StructureComposition(IonProperty(fast=True)), 
                                           StructureComposition(AtomicOrbitals())])
```

```python
#df_featurize_struc = structure_featurizer.featurize_dataframe(df, "structure_oxid")
df = structure_featurizer.featurize_dataframe(df, "structure_oxid", ignore_errors=True)
```
